```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns

from sklearn.cluster import KMeans
```

```python
image_as_array = mpimg.imread('../../Python for ML/DATA/palm_trees.jpg')
```

```python
image_as_array[:5]
```

```python
image_as_array.shape
```

```python
plt.figure(dpi=150)
plt.imshow(image_as_array)
```

```python
# (H,W,C) ==> (H*W, C)
```

```python
(h, w, c) = image_as_array.shape
```

```python
image_as_array2d = image_as_array.reshape(h * w, c)
```

```python
image_as_array2d
```

```python
model = KMeans(n_clusters=6)
```

```python
labels = model.fit_predict(image_as_array2d)
```

```python
labels
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int)
```

```python
rgb_codes
```

```python
rgb_codes[labels]
```

```python
q_img = np.reshape(rgb_codes[labels], (h,w, c))
```

```python
q_img
```

```python
plt.imshow(q_img)
```
