```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, RidgeCV, LassoCV, ElasticNetCV
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df = pd.read_csv('../../Python for ML/08-Linear-Regression-Models/Advertising.csv')
```

```python
df.head()
```

```python
X = df.drop('sales', axis=1)
```

```python
y = df['sales']
```

```python
polynomial_converter = PolynomialFeatures(degree=3, include_bias=False)
```

```python
poly_features = polynomial_converter.fit_transform(X, y)
```

```python
X.shape
```

```python
poly_features.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train[0]
```

```python
poly_features[0]
```

```python
ridge_model = Ridge(alpha=10)
```

```python
ridge_model.fit(X_train, y_train)
```

```python
test_predictions = ridge_model.predict(X_test)
```

```python
test_predictions
```

```python
MAE = mean_absolute_error(y_test, test_predictions)
```

```python
MAE
```

```python
RMSE = np.sqrt(mean_squared_error(y_test, test_predictions))
```

```python
RMSE
```

```python
ride_cv_model = RidgeCV(alphas=(0.1, 1.0, 10.0), scoring='neg_mean_absolute_error')
```

```python
ride_cv_model.fit(X_train, y_train)
```

```python
ride_cv_model.alpha_
```

```python
from sklearn.metrics import get_scorer_names
```

```python
get_scorer_names()
```

```python
test_predictions = ride_cv_model.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test, test_predictions)
```

```python
MAE
```

```python
RMSE = np.sqrt(mean_squared_error(y_test, test_predictions))
```

```python
RMSE
```

```python
ride_cv_model.coef_
```
