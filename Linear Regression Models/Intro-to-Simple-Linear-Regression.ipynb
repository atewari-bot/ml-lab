```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../../Python for ML/08-Linear-Regression-Models/Advertising.csv")
```

```python
df.head()
```

```python
df['total_spend'] = df['TV'] + df['radio'] + df['newspaper']
```

```python
df.head()
```

```python
sns.scatterplot(data=df, x='total_spend', y='sales')
```

```python
sns.regplot(data=df, x='total_spend', y='sales')
```

```python
X = df['total_spend']
y = df['sales']
```

```python
np.polyfit(X, y, deg=1)
```

```python
potential_spends = np.linspace(0, 500, 100)
```

```python
potential_sales = 0.04868788 * potential_spends + 4.24302822
```

```python
plt.plot(potential_spends, potential_sales)
```

```python
sns.regplot(data=df, x=df['total_spend'], y=df['sales'])
plt.plot(potential_spends, potential_sales, color='red')
```
