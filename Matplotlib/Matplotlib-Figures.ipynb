```python
# blank canvas
fig = plt.figure()

# Large axes
ax1 = fig.add_axes([0, 0, 1, 1])
ax1.plot(a, b)
ax1.set_xlim(0, 8)
ax1.set_ylim(0, 8000)
ax1.set_title('Power of 4')
ax1.set_xlabel('A')
ax1.set_ylabel('B')

# Small axes
ax2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])
ax2.plot(a, b)
ax2.set_xlim(1, 2)
ax2.set_ylim(0, 50)
ax2.set_title('Zoomed in')
ax2.set_xlabel('A')
ax2.set_ylabel('B')
```
